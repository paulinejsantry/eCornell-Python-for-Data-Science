## Exploring your own dataset

I will be exploring the 2023 NFL play by play data to predict if an offensive play will be a pass or rush.

Data source: https://nflsavant.com/about.php
Click "Download the 2023 play by play data"

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('pbp-2023.csv', header=0)
```

```python
#remove columns which are ALL NA values
df2 = df.dropna(how = 'all')
df2.head()
```

```python
#define the x and y variables
# had to remove Offensive & Defensive team since they are categorical variables

x_vars = [
    #""'OffenseTeam',
    # 'DefenseTeam',
    'Quarter',
    'Minute',
    
    'Down',
    'ToGo',
    'YardLine',
    'SeriesFirstDown',
]
y_vars = [
    'IsRush',
    'IsPass',
]
x = df2[x_vars]
y = df2[y_vars]
print(x.shape)
print(y.shape)
```

```python
# import 
from sklearn import svm, tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
# split dataset in to training & testing
datasubsets = train_test_split(x, y, random_state = 0)

(x_train, x_test, y_train, y_test) = datasubsets

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#build decision tree
clf1 = tree.DecisionTreeClassifier()
clf1.fit(x_train, y_train)
predictions1 = clf1.predict(x_test)
score1 = accuracy_score(y_test,predictions1)
print(score1)
```
